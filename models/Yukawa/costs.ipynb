```python
def n_terms_bdag_b_a_a(K):
    n_terms = 0

    for k in np.arange(5/2, np.floor(K - 1/2) + 1/2 + 1, 1):
        for n in np.arange(1, k - 3/2 + 1):
            n_terms += np.floor(k - n)
    return n_terms

def n_terms_bdag_b_adag_a(K):
    return np.sum([i**2 for i in np.arange(1, np.floor(K - 1/2) + 1/2)])

def n_terms_bdag_ddag_a_a(K):
    return np.sum([n * np.floor(n/2) for n in np.arange(2, np.floor(K) + 1)])

def n_terms_bdag_ddag_adag_a(K):
    n_terms = 0

    for s in np.arange(2, np.floor(K) + 1, 1):
        for n in np.arange(1, s - 1 + 1):
            n_terms += ((s - n))-(s-n)%2
    return n_terms
    
    
```

```python
def expected_n_terms(K, verbose = False):
    n_free = np.floor(K) + 2 * np.floor(K + 1/2)
    n_3pt = (
        2 * np.floor(K - 1/2) * (np.floor(K - 1/2) + 1) 
        + np.floor(K) * (np.floor(K) + 1) - 2 * np.floor((K + 1)/2)
    )
    n_4pt_type1 = 4 * n_terms_bdag_b_a_a(K)
    n_4pt_type2 = 2 * n_terms_bdag_b_adag_a(K)
    n_4pt_type3 = 2 * n_terms_bdag_ddag_a_a(K)
    n_4pt_type4 = 2 * n_terms_bdag_ddag_adag_a(K)
    
    if verbose:
        print(n_4pt_type1, n_4pt_type2, n_4pt_type3, n_4pt_type4, n_4pt_type1+ n_4pt_type2+ n_4pt_type3+ n_4pt_type4 )

    return n_free + n_3pt + n_4pt_type1+ n_4pt_type2 + n_4pt_type3 + n_4pt_type4

```

```python
def inst_cost(K):
    if K % 2 == 0:
        #2Z
        return (5/3) * K**3 - (2) * K**2 + (1/3) * K
    elif K % 2 == 1.0:
        #2Z + 1
        return (5/3) * K**3 - (2) * K**2 - (2/3) * K + 1
    elif K % 2 == 0.5:
        #2Z + 1/2
        return (5/3) * K**3 - (3/2) * K**2 + (7/12) * K - 1/8
    elif K % 2 == 1.5:
        #2Z + 3/2
        return (5/3) * K**3 - (3/2) * K**2 + (5/12) * K + 3/8
```

```python
def bare_cost(K_vals):
    costs = []

    for K in K_vals:
        free_cost = np.floor(K) + 2 * np.floor(K + 1/2)
        cost_3pt = 2 * np.floor(K - 1/2) * (np.floor(K - 1/2) + 1) + np.floor(K) * (np.floor(K) + 1) - 2 * np.floor((K + 1)/2)
        cost_inst = inst_cost(K)
        costs.append(free_cost + cost_3pt + cost_inst)

    return np.array(costs)
```

```python
def boson_ex_cost(K):
    if K % 2 == 0:
        #2Z
        return (5/6) * K**3  - (5/3) * K
    elif K % 2 == 1.0:
        #2Z + 1
        return (5/6) * K**3  - (1/3) * K + 1/2
    elif K % 2 == 0.5:
        #2Z + 1/2
        return (5/6) * K**3 - (1/4) * K**2 - (5/24) * K + 1/16
    elif K % 2 == 1.5:
        #2Z + 3/2
        return (5/6) * K**3 - (1/4) * K**2 - (17/24) * K - 3/16


def fermion_ex_cost(K):
    if K % 2 == 0:
        #2Z
        return (5/3) * K**3 - (2) * K**2 + (1/3) * K
    elif K % 2 == 1.0:
        #2Z + 1
        return (5/3) * K**3 - (2) * K**2 - (2/3) * K + 1
    elif K % 2 == 0.5:
        #2Z + 1/2
        return (5/3) * K**3 - (3/2) * K**2 + (7/12) * K - 1/8
    elif K % 2 == 1.5:
        #2Z + 3/2
        return (5/3) * K**3 - (3/2) * K**2 + (5/12) * K + 3/8
```

```python
for k in np.arange(1.5, 1e4, 0.5):
    if inst_cost(k) != fermion_ex_cost(k):
        print(k)
```

```python
def effective_cost(Kvals):
    costs = []

    for K in Kvals:
        free_cost = (np.floor(K) + 2 * np.floor(K + 1/2)) 
        cost_3pt = 2 * np.floor(K - 1/2) * (np.floor(K - 1/2) + 1) + np.floor(K) * (np.floor(K) + 1) - 2 * np.floor((K + 1)/2)
        cost_inst = inst_cost(K)

        cost_boson_ex = boson_ex_cost(K)
        # cost_fermion_ex = fermion_ex_cost(K)
        
        costs.append(free_cost + cost_3pt + cost_inst + cost_boson_ex)

    return np.array(costs)
```

```python
def b_exch_cos(Kvals):
    costs = []

    for K in Kvals:

        cost_boson_ex = boson_ex_cost(K)
        # cost_fermion_ex = fermion_ex_cost(K)
        
        costs.append(cost_boson_ex)

    return np.array(costs)
```
