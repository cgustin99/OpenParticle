```python
#Imports
from openparticle import Fock, ParticleOperator
from openparticle.qubit_mappings import *
from openparticle import FermionOperator, AntifermionOperator, BosonOperator
from openparticle import ConjugateFock
```

```python
#An example state

f_occ = [2] #One fermion in mode 2
af_occ = [1, 3] #One antifermion in mode 1, one in mode 3
b_occ = [(0, 3)] #Three bosons in mode 0

x1 = Fock(f_occ = f_occ, af_occ = af_occ, b_occ = b_occ)
x2 = Fock(f_occ = f_occ, af_occ = af_occ, b_occ = b_occ)
x3 = Fock(f_occ = f_occ, af_occ = [2, 3], b_occ = b_occ)
print(x1)
print(x1 + x2)
print(x1 + x3)

x1_c = x1.dagger()
x2_c = x2.dagger()
x3_c = x3.dagger()
print(x2_c + x1_c)
print(x2_c + x3_c)

sum = x1 + x2
print(sum + x3)
print(x3 + sum)

conjsum = x2_c + x3_c
print(conjsum)
print(conjsum + x3_c)
```
